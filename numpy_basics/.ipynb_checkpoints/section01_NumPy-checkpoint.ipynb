```python
# This exercise demonstrates how to use the np.where() function to filter data based on a condition. In this case, we identify the indices of all individuals who are considered adults (18 years or older). The function np.where(ages >= 18) evaluates which elements in the ages array meet the condition (ages >= 18). The function returns a tuple of arrays, and the first array contains the indices of these elements. This allows us to easily locate which entries in the original array meet our condition, which is a common task in data analysis when segmenting or filtering data based on criteria.

ages = np.array([25, 12, 15, 64, 35, 80, 45, 10, 22, 55])

adult_ages = np.where(ages >= 18)
 
# Output the indices
print("Ages of adults:", adult_ages)
```

# Assignment 6: Aggregating and Sorting Arrays

This exercise helps you learn how to sort arrays directly in memory with the .sort() method, altering the original data structure, and then directly access the sorted data. After using .sort(), the ages are arranged from the youngest to the oldest, allowing easy retrieval of the minimum and maximum ages. This method is useful for operations requiring rearranged data without preserving the original order, making data analysis workflows both efficient and straightforward.

```python
ages = np.array([25, 12, 15, 64, 35, 80, 45, 10, 22, 55])


# Sort the array in-place
new_array = np.sort(ages)

# Output the sorted array
print (new_array)


# Retrieve and output the minimum and maximum age

min_age = new_array.min()
max_age = new_array.max()
print (min_age)
print (max_age)
```

# Assignment 7: Bringing it All Together

Ok, final NumPy task - let's read in some data with the help of Pandas.

Our data scientist provided the code to read in a csv as a Pandas dataframe, and has converted the two columns of interest to arrays.

* Filter `sales_array` down to only sales where the product family was produce. 

* Then, randomly sample roughly half (random number < .5) of the produce sales and report the mean and median sales. Use a random seed of 2022.

* Finally, create a new array that has the values 'above_both', 'above_median', and 'below_both' based on whether the sales were above the median and mean of the sample, just above the median of the sample, or below both the median and mean of the sample. 

```python
import pandas as pd
import numpy as np

retail_df = pd.read_csv(
    "../retail/retail_2016_2017.csv", skiprows=range(1, 11000), nrows=1000
)

family_array = np.array(retail_df["family"])
sales_array = np.array(retail_df["sales"])
#print(sales_array.mean())
```

```python
produce_array = sales_array [family_array == "PRODUCE"]

rng = np.random.default_rng(2022)
randaom_array = rng.random(30)
sampled_array = produce_array [randaom_array < .5]
```

```python
mean = sampled_array.mean()
print (mean)
```

```python
median = np.median(sampled_array)
print (median)
```

```python
np.where(

    sampled_array < median,
    "below_both",
    np.where(
        sampled_array > mean, "above_both", "above_median"
    )
)
```
