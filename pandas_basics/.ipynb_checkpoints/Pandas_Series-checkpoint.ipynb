# Assignment 1: Series Basics

The code has been previded to create an array, `oil_array` from a dataframe column. 

* Convert `oil_array` into a Pandas Series, called `oil_series`. Give it a name!
* Return the name, dtype, size, and index of `oil_series`.

Take the mean of the values array. 

Then, convert the series to integer datatype and recalculate the mean. 


```python
# create a DataFrame from the oil file, drop missing values
oil = pd.read_csv("../retail/oil.csv").dropna()

# Grab 100 rows of oil prices
oil_array = np.array(oil["dcoilwtico"].iloc[1000:1100])
oil_series = pd.Series(oil_array, name= "test")
print(oil_series.name)
print(oil_series.dtype)
print(oil_series.index)
print(oil_series.values.mean())
oil_series = oil_series.astype("int64")
print(oil_series.values.mean())
print(oil_series.dtype)
oil_series = oil_series.astype("float")
```

```python
#In this exercise, you dealt with a series of age data that included missing values and an outlier (age 110). You first used .isna.sum() to determine that there were three missing values in your dataset. Calculating the median with the .median() method, you included all values, even the outlier, because the median is robust to extreme values—unlike the mean, which could have been significantly skewed by the age of 110. This makes the median a safer choice for imputation in such scenarios.

#This method of imputation is particularly useful in real-world data settings where data may not be perfectly clean or uniformly distributed. The filled dataset is more complete and can be used for further analysis without the bias that an extreme value or missing data might introduce

age_data = pd.Series([25, np.nan, 30, 45, 110, 37, np.nan, 42, np.nan, 52])
 
# Calculate the number of missing values
missing_count = age_data.isna().sum()
print("Number of missing values:", missing_count)
 
# Calculate the median of the age data including the outlier
median_age = age_data.median()
 
# Fill missing values with the median age
age_data_filled = age_data.fillna(median_age)
 
# Output the result
print("Data after median imputation:")
print(age_data_filled)
```

```python
## This exercise demonstrates the practical application of conditional logic using Pandas. By employing the  .where() method with a logical inversion, you efficiently classify the ages into 'Adult' and 'Teen'. This method is not only concise but also showcases a powerful 
#feature of Pandas that can simplify complex logical operations in data processing tasks.

ages = pd.Series([16, 21, 15, 25, 14, 19, 22, 17, 13, 20])

# Use the Pandas .where() function to classify ages
# Anyone 18 or older is classified as "Adult", otherwise "Teen"
age_groups = ages.where(
    ages < 18, "Adult").where(ages > 18, "Teen" )
```
